```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower();    # columns convert into lower case 
df.columns = df.columns.str.replace(' ', '_');   # rolumns replace with spaces into 
```

```python
df.isnull().sum()
```

```python
df = df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'});
```

```python
df.columns
```

```python
# create a columns age_group 
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels);
```

```python
df[['age','age_group']].head(10);
```

```python
df.head()
```

```python
# Create Columns Purchase Frequency Days

frequency_mapping = {
    'Fortnightly':14,
    'Weekly':7,
    'Monthly':30,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Annually':365,
    'Every 3 Months':90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping);
```
